```python
import yaml
yaml_file_path = '/content/dataset/data.yaml'

with open(yaml_file_path, 'r') as f:
    data_config = yaml.safe_load(f)

data_config['path'] = '/content/dataset'

with open(yaml_file_path, 'w') as f:
    yaml.safe_dump(data_config, f)

print(f"Updated {yaml_file_path} with path: {data_config['path']}")
```

```python
import os

train_imgs = len(os.listdir('/content/dataset/images/train'))
val_imgs = len(os.listdir('/content/dataset/images/val'))
test_imgs = len(os.listdir('/content/dataset/images/test'))

print("Train:", train_imgs)
print("Val:", val_imgs)
print("Test:", test_imgs)
print("Total:", train_imgs + val_imgs + test_imgs)
```
